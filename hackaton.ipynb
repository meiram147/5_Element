```python
import numpy as np
np.random.seed(1)
```

```python
import openpyxl

wookbook1 = openpyxl.load_workbook("/Users/ayanryskulov/Documents/hackaton/boolean.xlsx")
worksheet1 = wookbook1.active
boolean_input_data = []
boolean_result = []
row_num = 70

for i in range(0, row_num):
    count = 0
    input_row = []
    for col in worksheet1.iter_cols(0, worksheet1.max_column):
        if count < 4:
            input_row.append(col[i].value)
            count += 1
        else:
            boolean_result.append(col[i].value)
    boolean_input_data.append(input_row)
```

```python
def relu(x):
    return (x > 0) * x

def relu2deriv(output): 
    return output > 0 
alpha = 0.2 
hidden_size = 5
```

```python
data = np.array(boolean_input_data)
result = np.array(boolean_result).T
```

```python
weights_0_1 = 2 * np.random.random((4, hidden_size)) - 1 
weights_1_2 = 2 * np.random.random((hidden_size, 1)) - 1
```

```python
for iteration in range(60): 
    layer_2_error = 0
    
    for i in range(len(data)):
        layer_0 = data[i:i+1]
        layer_1 = relu(np.dot(layer_0, weights_0_1)) 
        layer_2 = np.dot(layer_1, weights_1_2)
        layer_2_error += np.sum((layer_2 - result[i:i+1]) ** 2)
        layer_2_delta = (layer_2 - result[i:i+1]) 
        layer_1_delta = layer_2_delta.dot(weights_1_2.T) * relu2deriv(layer_1)
        weights_1_2 -= alpha * layer_1.T.dot(layer_2_delta) 
        weights_0_1 -= alpha * layer_0.T.dot(layer_1_delta)
    
    if(iteration % 10 == 9):
        print("Error: " + str(layer_2_error))
```

```python
result_g = []
for i in range(70):
    test = np.array(boolean_input_data[i])
    layer_0 = test
    layer_1 = np.dot(layer_0, weights_0_1) 
    layer_1 = relu(layer_1)
    layer_2 = np.dot(layer_1, weights_1_2)
    if layer_2 > 0.5:
        result = True
    else:
        result = False
    result_g.append(result)
        
```

```python
import pandas as pd

df = pd.DataFrame({'Actual': boolean_result, 'Predicted': result_g})
df
```
