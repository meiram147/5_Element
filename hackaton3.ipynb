```python
import numpy as np
import openpyxl
np.random.seed(1)
```

```python
def relu(x):
    return (x > 0) * x

def sig(x):
    return 1/(1 + np.exp(-x))


def relu2deriv(output): 
    return output > 0 
alpha = 0.4
hidden_size = 12
```

```python
wookbook2 = openpyxl.load_workbook("integer.xlsx")

worksheet2 = wookbook2.active

int_input_data = []
int_result = []
row_num = 63
division = 10000
for i in range(0, row_num):
    count = 0
    input_row = []
    for col in worksheet2.iter_cols(0, worksheet2.max_column):
        if count < 4:
            input_row.append(col[i].value / division)
            count += 1
        else:
            int_result.append(col[i].value / division)
    int_input_data.append(input_row)
    
```

```python
data = np.array(int_input_data)
result = np.array(int_result).T
```

```python
weights_0_1 = 0.2 * np.random.random((4, 12)) - 0.1 
weights_1_1 = 0.2 * np.random.random((12, 12)) - 0.1
weights_1_2 = 0.2 * np.random.random((12, 1)) - 0.1
```

```python
for iteration in range(22000): 
    layer_2_error = 0
    
    for i in range(len(data)):
        layer_0 = data[i:i+1]
        layer_1 = sig(np.dot(layer_0, weights_0_1)) 
        layer_1_2 = sig(np.dot(layer_1, weights_1_1))
        layer_2 = np.dot(layer_1_2, weights_1_2)
        
        
        layer_2_error += np.sum((layer_2 - result[i:i+1]) ** 2)
        
        layer_2_delta = (layer_2 - result[i:i+1]) 
        
        layer_1_2_delta = layer_2_delta.dot(weights_1_2.T)
        
        layer_1_delta = layer_1_2_delta.dot(weights_1_1.T)
        
        weights_1_2 -= alpha * layer_1.T.dot(layer_2_delta) 
        weights_1_1 -= alpha * layer_1_2.T.dot(layer_1_2_delta) 
        weights_0_1 -= alpha * layer_0.T.dot(layer_1_delta)
    
    if(iteration % 100 == 9):
        print("Error: " + str(layer_2_error) + "   " + "Iter " + str(iteration))
```

```python
num = 18

test = np.array(int_input_data[num])
layer_0 = test
layer_1 = np.dot(layer_0, weights_0_1) 
layer_1 = sig(layer_1)
layer_1_2 = np.dot(layer_1, weights_1_1) 
layer_1_2 = sig(layer_1_2)
layer_2 = np.dot(layer_1_2, weights_1_2)


print(layer_2 * 10000)
print(int_input_data[num])
```
